## 3. Traitement des colonnes 

### 3.1 colonne price 

```python
list_of_price = []
for value in df1['price'] : 
    value = value.replace('$', '')
    value = value.replace(',', '')
    list_of_price.append(value)
    print(value)
```

```python
df1['price'] = list_of_price
```

```python
print(df1['price'].dtype)
```

la colonne price est en object elle est pas exploitable on va donc la transformer en int

```python
df1['price'] = df1['price'].astype('float32')
```
